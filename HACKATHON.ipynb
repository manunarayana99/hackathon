```python
# DATA CLEANING & PREPROCESSING
# STEP-1 RATING COLUMMN (RATE)
df["rate"]=df["rate"].fillna("-")
df["rate"]
```

```python
# REmove /5 and retain only numeric value
df["rate"]=df["rate"].str.rstrip("/5")
df["rate"]
```

```python
# convert to float using pd.to_numeric()
df["rate"]=pd.to_numeric(df["rate"],errors="coerce")
df["rate"]
```

```python
df.describe()
```

```python
# fill missing values using the median rating
df["rate"]=df["rate"].fillna(df["rate"].median())
df["rate"]
```

```python
# STEP 2 - COST COLUMN 
# REMOVE COMMAS FROM NUMERIC STRINGS
df["approx_costfor_two_people"]=df["approx_costfor_two_people"].str.replace(",","")
df["approx_costfor_two_people"]
```

```python
# CONVERT TO NUMERIC
df["approx_costfor_two_people"]=pd.to_numeric(df["approx_costfor_two_people"])
df["approx_costfor_two_people"]
```

```python
# FILL MISSING VALUES WITH MEDIAN COST
df["approx_costfor_two_people"]=df['approx_costfor_two_people'].fillna(df["approx_costfor_two_people"].median())
df["approx_costfor_two_people"]
```

```python
# STEP -3 CATEGORICAL COLUMNS
# dish_liked → Replace NaN with "Not Available"
df["dish_liked"]=df["dish_liked"].fillna("Not Available")
df["dish_liked"]
```

```python
# cuisines → Replace NaN with "Other"
df['cuisines']=df["cuisines"].fillna("Other")
df["cuisines"]
```

```python
# rest_type → Replace NaN with "Unknown"
df['rest_type']=df["rest_type"].fillna("Unknown")
df["rest_type"]
```

```python
df["rest_type"].isnull().sum()
```

```python
# STEP 4- VOTES COLUMN
# Fill missing values in votes with median
df["votes"]=df["votes"].fillna(df["votes"].median())
df["votes"]
```

```python
df['votes'].isnull().sum()
```

```python
# Step 5: Binary Encoding
# Convert the following binary fields:
df["online_order"]=df["online_order"].map({"Yes":1,"No":0})
df["online_order"]
```

```python
df["book_table"]=df["book_table"].map({"Yes":1,"No":0})
df["book_table"]
```

```python
# Step 6: Data Type Conversion
# Ensure the following conversions:
# rate → float , votes → integer , approx_costfor_two_people → integer
print(df["rate"].dtype)
print(df["votes"].dtype)
print(df["approx_costfor_two_people"].astype("int").dtype)
```

```python
# 2. How many restaurants serve North Indian cuisine?
north=df[df["cuisines"].str.contains("North Indian",case=False,na=False)]
len(north)
```

```python
df.columns
```

```python
# 3. What cuisine is most commonly offered by restaurants in Bangalore?
cuisines = df['cuisines'].str.split(', ')
all = [i for sublist in cuisines for i in sublist]
counts = Counter(all)
common = counts.most_common(1)[0][0]
common
```

```python
# 4. Which locality in Bangalore has the highest average cost for dining (for two people)?
avg = df.groupby('listed_incity')['approx_costfor_two_people'].mean()
highest_avg= avg.idxmax()
highest_avg
```

```python
# 5. Which restaurant type has the top rating with over 1000 votes
over_df = df[df['votes'] > 1000]
sorted_df = over_df.sort_values(by=['rate'], ascending=False)
top= sorted_df.iloc[0]['rest_type']
top
```

```python
#6. How much does it cost at minimum to eat out in Bangalore?

min_cost=df["approx_costfor_two_people"].min()
print("minimum cost= ",min_cost)
print("average cost =",df["approx_costfor_two_people"].mean())
```

```python
#7. What percentage of total online orders is received by restaurants in Banashankari?
banashankari = df[df['listed_incity'] == 'Banashankari']
banashankari_online = banashankari[banashankari['online_order'] == 1]

banashankari_online_count = len(banashankari_online)
total_banashankari_order_count = len(banashankari)
percentage = (banashankari_online_count / total_banashankari_order_count) * 100
percentage
```

```python
# 8. Which locality has the most restaurants with over 500 votes and a rating below 3.0?
votedrate_df = df[(df['votes'] > 500) & (df['rate'] < 3.0)]
restcount= votedrate_df.groupby('listed_incity').size() 
most = restcount.idxmax()
most
```

```python
#9. Which locality in Bangalore should Zomato target for expansion based on restaurant type diversity?

rest_type = df.groupby('listed_incity')['rest_type'].nunique()
target_locality = rest_type.idxmax()
target_locality
```

```python
# 10. What's the average cost difference between buffet and delivery restaurants?

buffet_restaurants = df[df['rest_type'] == 'Buffet']
delivery_restaurants = df[df['rest_type'] == 'Delivery']
avg_buffet_cost = buffet_restaurants['approx_costfor_two_people'].mean()
avg_delivery_cost = delivery_restaurants['approx_costfor_two_people'].mean()
cost_difference = avg_buffet_cost - avg_delivery_cost

cost_difference
```

```python
# 11 . What is the maximum number of votes received by any restaurant with online ordering?

online= df[df['online_order'] == 1]
max_votes = online['votes'].max()
max_votes
```

```python
# 12. What is the average rating of restaurants that serve both North Indian and Chinese cuisines?

both= df[df['cuisines'].str.contains('North Indian', case=False) & df['cuisines'].str.contains('Chinese', case=False)]
avg_rating = both['rate'].mean()
print("The average rating of restaurants that serve both = ",avg_rating)
```

```python
# 13. What is the most profitable area for Zomato based on potential revenue estimation?

total = df.groupby('listed_incity')['votes'].sum()
most_profit= total.idxmax()

print("The most profitable area for Zomato =" ,most_profit)
```

```python
# 14. If Zomato wants to reduce customer complaints, which restaurant type should they focus on?

rest_type_stats = df.groupby('rest_type').agg({'rate': 'mean', 'votes': 'sum'})
sign_votes = rest_type_stats[rest_type_stats['votes'] > 1000]
to_focus = sign_votes['rate'].idxmin()
print("Zomato should focus on the = ",to_focus)
```

```python
# 15. In which area should Zomato invest by considering high rating (rate > 4.2), high number of votes (> 500) and including online orders?

df3 = df[(df['rate'] > 4.2) & (df['votes'] > 500) & (df['online_order'] == 1)]
restaurant_count =df3.groupby('listed_incity')['rest_type'].count()  
target_locality = restaurant_count.idxmax()
print("Zomato should invest in =",target_locality)
```
